### 손실 함수(Loss Function)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = np.array([0,1,0,0])
yhat1 = np.array([0.1,0.7,0.2,0])
yhat2 = np.array([0.1,0.2,0.7,0])
N = y.size
```

- 평균 제곱 오차(MSE: Mean Squared Error)

```python
mse1 = np.sum(np.square(yhat1-y)) / N
mse2 = np.sum(np.square(yhat2-y)) / N

print(mse1, mse2)
```

- 평균 제곱근 오차(RMSE: Root MSE)

```python
rmse1 = np.sqrt(mse1)
rmse2 = np.sqrt(mse2)

print(rmse1, rmse2)
```

- 평균 절대오차(MAE : Mean Absolute Error)

```python
mae1 = np.sum(np.abs(yhat1 - y)) / N
mae2 = np.sum(np.abs(yhat2 - y)) / N
print(mae1, mae2)
```

- Categorical Cross Entropy

```python
def cross_entropy_error(yhat, y):
    if yhat.ndim == 1:
        yhat = yhat.reshape(1,-1)
    if y.ndim == 1:
        y = y.reshape(1,-1)

    if yhat.size == y.size:
        y = y.argmax(axis=1)
    
    batch_size = yhat.shape[0]
    return -np.sum(np.log(yhat[np.arange(batch_size), y])) / batch_size
```

```python
ccel1 = cross_entropy_error(yhat1, y)
ccel2 = cross_entropy_error(yhat2, y)

print(ccel1, ccel2)
```
