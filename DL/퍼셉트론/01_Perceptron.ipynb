## Perceptron

```python
import numpy as np
```

### AND Gate

```python
weight = np.array([1,1])
bias = -1

for x1, x2 in [(0,0),(0,1), (1,0), (1,1)]:
    x = np.array([x1,x2])
    sum = np.dot(weight, x) + bias
    print(0 if sum <= 0 else 1)
```

```python
def AND_gate(x1,x2):
    weight = np.array([1,1])
    bias = -1
    
    x = np.array([x1,x2])
    sum = np.dot(weight, x) + bias
    return 0 if sum <= 0 else 1
```

```python
for x1, x2 in [(0,0),(0,1),(1,0),(1,1)]:
    print(AND_gate(x1,x2))
```

### OR Gate

```python
def OR_gate(x1,x2):
    weight = np.array([2,2])
    bias = -1

    x = np.array([x1,x2])
    sum = np.dot(weight, x) + bias

    return 0 if sum <= 0 else 1
```

```python
for x1, x2 in [(0,0),(0,1),(1,0),(1,1)]:
    print(OR_gate(x1,x2))
```

### NAND Gate

```python
def NAND_gate(x1,x2):
    weight = np.array([-1,-1])
    bias = 2

    x = np.array([x1,x2])
    sum = np.dot(weight, x) + bias

    return 0 if sum <= 0 else 1
```

```python
for x1, x2 in [(0,0),(0,1),(1,0),(1,1)]:
    print(NAND_gate(x1,x2))
```
