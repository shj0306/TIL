- MLP

```python
import numpy as np
```

```python
def AND_gate(x1,x2):
    weight = np.array([1,1])
    bias = -1
    
    x = np.array([x1,x2])
    sum = np.dot(weight, x) + bias
    return 0 if sum <= 0 else 1

def OR_gate(x1,x2):
    weight = np.array([2,2])
    bias = -1

    x = np.array([x1,x2])
    sum = np.dot(weight, x) + bias

    return 0 if sum <= 0 else 1

def NAND_gate(x1,x2):
    weight = np.array([-1,-1])
    bias = 2

    x = np.array([x1,x2])
    sum = np.dot(weight, x) + bias

    return 0 if sum <= 0 else 1
```

```python
n1, n2 = [], []
for x1, x2 in [(0,0),(0,1),(1,0),(1,1)]:
    print(NAND_gate(x1,x2))
```

- XOR gate

```python
def XOR_gate(x1, x2):
    return AND_gate(NAND_gate(x1,x2), OR_gate(x1,x2))
```

```python
print('x1,x2\tn1,n2\tXOR')
for x1, x2 in [(0,0), (0,1), (1,0), (1,1)]:
    print(f'{x1:2d},{x2:2d}\t{NAND_gate(x1,x2):2d},{OR_gate(x1,x2):2d}\t{XOR_gate(x1,x2):3d}')
```
